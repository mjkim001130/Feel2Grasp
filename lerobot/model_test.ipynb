```python
from lerobot.datasets.lerobot_dataset import LeRobotDataset
repo_id = "mjkim00/Feel2Grasp"

ds = LeRobotDataset(repo_id)
print("dataset root =", ds.root if hasattr(ds,"root") else ds.meta.root)
```

```python
from pathlib import Path
import pyarrow.dataset as ds

episodes_dir = Path("/home/joy4mj/.cache/huggingface/lerobot/mjkim00/Feel2Grasp/meta/episodes")
print("episodes_dir exists:", episodes_dir.exists())

# 재귀로 parquet를 찾는지
files = list(episodes_dir.rglob("*.parquet"))
print("num parquet =", len(files))
print("first few =", files[:3])

# pyarrow가 dataset으로 인식하는지
dataset = ds.dataset(str(episodes_dir), format="parquet")
print("schema =", dataset.schema)
print("num fragments =", len(dataset.get_fragments()))
```

```python
import pyarrow.parquet as pq

p = "/home/joy4mj/.cache/huggingface/lerobot/mjkim00/Feel2Grasp/meta/episodes/chunk-000/file-000.parquet"
pf = pq.ParquetFile(p)
print("rows =", pf.metadata.num_rows)
print("schema =", pf.schema)
```
